```python
import yaml

# it read the existing data.yaml
with open('/content/data.yaml', 'r') as file:
    data = yaml.safe_load(file)

# Update the root path to Colab's local storage
data['path'] = '/content/yolo_dataset'

# Write it back
with open('/content/data.yaml', 'w') as file:
    yaml.dump(data, file)
```

```python
!pip install ultralytics -q
from ultralytics import YOLO

# it loads the pretrained yolov8 nano model
model = YOLO('yolov8n.pt')

results = model.train(
    data='/content/data.yaml',
    epochs=50,
    imgsz=640,
    batch=16,
    device=0, # 0 as we are doing training on google colab
    project='/content/drive/MyDrive/Runs', # saving results on my googledrive
    name='exdark_baseline_1',
    plots=True
)
```
