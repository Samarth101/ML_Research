```python
from ultralytics import YOLO
```

```python
def zero_shot_visuals():
    print("Loading RAW off-the-shelf YOLOv8 Model (COCO Pre-trained)...")
    model = YOLO('yolov8n.pt') # The standard, un-fine-tuned model

    # 1. Test on Original Dark Images
    print("\nRunning inference on DARK images...")
    model.predict(
        source='yolo_dataset/images/val', 
        save=True,
        conf=0.25,
        imgsz=640,
        device='mps', 
        name='zeroshot_dark_test'
    )

    # 2. Test on GAN-Enhanced Bright Images
    print("\nRunning inference on BRIGHTENED images...")
    model.predict(
        source='yolo_dataset/images/val_enhanced', 
        save=True,
        conf=0.25,
        imgsz=640,
        device='mps', 
        name='zeroshot_bright_test'
    )
    print("\nDone! Open 'runs/detect/zeroshot_dark_test' and compare them side-by-side with 'runs/detect/zeroshot_bright_test'.")
```

```python
if __name__ == '__main__':
    zero_shot_visuals()
```
