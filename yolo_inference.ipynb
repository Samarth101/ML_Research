```python
from ultralytics import YOLO
```

```python
def evaluate_baseline():
    print("Loading baseline yolov8 model..")

    model_path="Runs/exdark_baseline_1/weights/best.pt"
    dataset_yaml="data.yaml"

    model=YOLO(model_path)

    model.predict(
        source="yolo_dataset/images/val",
        conf=0.25,
        imgsz=640,
        save=True,
        device="mps",
        name="baseline_visual_test"
    )


    print("mAP METRICS : \n\n")
    metrics=model.val(
        split='val',
        data=dataset_yaml,
        device="mps"
    )

    print(f"mAP@0.5 (Standard): {metrics.box.map50:.4f}")
    print(f"mAP@0.5:0.95 (Strict): {metrics.box.map:.4f}")
```

```python
if __name__ == '__main__':
    evaluate_baseline()
```
